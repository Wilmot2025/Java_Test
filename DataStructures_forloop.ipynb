# List Functions

```python
# adding elements to the list
num=[12,14,15,2,3,8]
# add 10 to this list
num.append(10)
print(num)
# insert an element at some specific position
num.insert(2,17)
print(num)
# adding more than one element
num.extend([0,4])
print(num)
# adding multiple elements at some position
num[3:3]=[20,30,25]
print(num)
```

```python
# removing the elements
num=[12,14,15,2,3,8,2]
num.remove(2) #remove first occurence of 2
print(num)
num.pop(2) #removes the last element
print(num)
num.clear() #removes all elements
print(num)
```

```python
num=[12,14,15,2,3,8,2]

del num[1:3]
num
```

```python
num=[12,14,15,2,3,8,2]

print(num.index(14)) #index of first occurence
print(num.count(2))
```

# for loop

```python
# for item in sequence:
    # perform the operation on item
```

```python
fruit=["apple","banana","mango"]
for i in fruit:
  print(i)
```

```python
fruit=["apple","banana","mango"]
for i in fruit:
  if i=="banana":
      print(i)
```

```python
for i in range(5):
  print(i)
```

```python
# range(5)---[0,1,2,3,4]
# range(1,5)---[1,2,3,4]
# range(1,5,2)---[1,3]
# range(start at:stop before:step size)
```

```python
fruit=["apple","banana","mango"]
for i in fruit:
  print(f"the length of {i} is {len(i)}")
```

```python
fruit=["apple","banana","mango"]
l=len(fruit)  #l=3
for i in range(l):
  print(i, fruit[i])
```

```python
# create a list of squares of first 10 whole numbers
sq=[]
print(f"before entering the loop {sq}")
for i in range(10):
  square=i**2
  sq.append(square)
  print(f"At step {i+1} value of list is {sq}")
```

```python
# create a list with double of numbers from 1 to 10
double=[]
for i in range(1,11):
  double.append(i*2)
print(double)
```

```python
# list comprehension
# short way of creating a list
sqr=[i**2 for i in range(10)]  #take the value of  from 0 t0 9 and square it and add it to the list
sqr
```

```python
sqr_even=[i**2 for i in range(10) if i%2==0]    #take the value of  from 0 t0 9 check if it is even and square it and add it to the list
sqr_even
```

```python
fruit=["apple","banana","mango"]
upper1=[i.upper() for i in fruit]
upper1
```
